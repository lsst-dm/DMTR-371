### Test case LVV-T2697: Verify implementation of Catalog Data Product Access

Verify that available catalog data products can be listed and retrieved.

```python
from lsst.daf.butler import Butler

import matplotlib.pyplot as plt
import numpy as np
import lsst.afw.display as afwDisplay
```

#### Initialize the butler pointing to DP0.2 data:

```python
# For DP0.2 data on the IDF:                                                                                                           
config = 'dp02'
collection = '2.2i/runs/DP0.2'
butler = Butler(config, collections=collection)
```

#### Define some data dimensions to use for querying datasets:

```python
tract = 3828
patch = 42
detector = 42
```

#### Execute a butler registry query to list `src` catalogs that match the selected dimensions.

Note that `src` catalogs (which are derived from `calexp` images) are defined with dimensions of visit and detector, but here we supply tract and patch constraints. This demonstrates that the query is able to identify which visits overlap the patch/tract combination that has been provided.

```python
data_refs = butler.registry.queryDatasets(
    datasetType="src",
    where=f"tract={tract} and patch={patch} and detector={detector} and skymap='DC2'")

data_ids = []

for data_ref in data_refs:
    print(data_ref.dataId.full)
    data_ids.append(data_ref.dataId.full)
```

This demonstrates the ability to identify and list available catalog datasets via the Butler middleware.

#### Select and retrieve a single "src" catalog for examination

```python
# extract the src catalog
src = butler.get('src', dataId=data_ids[0])
```

#### Examine the catalog to confirm that it is well-formed

```python
src.asAstropy()
```

```python
# Check the schema detailing the contents of the catalog:
print("The catalog has ", len(src.schema.getNames()), " columns.")

print("\n Catalog schema: ")
src.schema
```

The catalog looks reasonable. It contains measured positions, fluxes, shapes, and flags.
